#### Colllaborative Filtering

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Check for negative values in 'ReviewId' column
negative_review_ids = sampled_df[sampled_df['ReviewId'] < 0]

# Check for negative values in 'RecipeId' column
negative_recipe_ids = sampled_df[sampled_df['RecipeId'] < 0]

if not negative_review_ids.empty:
    print("Negative values found in 'ReviewId' column:")
    print(negative_review_ids)

if not negative_recipe_ids.empty:
    print("\nNegative values found in 'RecipeId' column:")
    print(negative_recipe_ids)
```

```python
# Creating a user-item matrix
user_item_matrix = sampled_df.pivot_table(index='ReviewId', columns='RecipeId', values='AggregatedRating', fill_value=0)
 # Filling missing values with 0
```

```python
# Spliting the user-item matrix into training and testing sets
train_matrix, test_matrix = train_test_split(user_item_matrix, test_size=0.2)
```

```python
print(train_matrix.shape)
```

```python
# Creating a Collaborative Filtering Model based on reviews by users

# Calculate user similarities using cosine similarity
user_similarities = cosine_similarity(train_matrix)
```

```python
# Make recommendations for a user based on similar users
def collaborative_filtering_user_based(user_id, num_recommendations):
    user_vector = train_matrix.loc[user_id].values.reshape(1, -1)
    user_sim_scores = user_similarities[user_id]
    weighted_sum = user_sim_scores.dot(train_matrix.values)
    predicted_ratings = weighted_sum / user_sim_scores.sum()
    
    # Exclude recipes the user has already interacted with
    user_interactions = train_matrix.loc[user_id]
    predicted_ratings[user_interactions > 0] = 0
    
    # Get the indices of top recommendations
    top_indices = predicted_ratings.argsort()[::-1][:num_recommendations]
    top_recommendations = train_matrix.columns[top_indices]
    
    return top_recommendations
```

```python
# Display top recommendations for a specific user
user_id = 1203093  
num_recommendations = 5  

# Get collaborative filtering recommendations
top_recommendations = collaborative_filtering_user_based(user_id, num_recommendations)

# Print or use the recommendations
print("Top Recommendations:")
print(top_recommendations)

```

#### Natural Language Processing (NLP)

Text Preprocessing (Sample for Ingredients)

```python
import nltk
nltk.download('stopwords')
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
# Sample text preprocessing for ingredients
def preprocess_text(text):
    # Tokenization and lowercasing
    words = re.findall(r'\w+', text.lower())
    
    # Removing stopwords
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]
    
    # Stemming
    stemmer = PorterStemmer()
    stemmed_words = [stemmer.stem(word) for word in filtered_words]
    
    return ' '.join(stemmed_words)
```

```python
# Apply preprocessing to the 'RecipeIngredientParts' column
sampled_df.loc[:, 'RecipeIngredientParts'] = sampled_df['RecipeIngredientParts'].apply(preprocess_text).copy()
```

Text-Based Recommendations (Using TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sampled_df['RecipeIngredientParts'])
```

```python
# Calculate cosine similarity between recipes based on ingredients
ingredient_similarities = cosine_similarity(tfidf_matrix)
```

```python
# Make recommendations based on ingredient similarities
def text_based_recommendations(recipe_index, num_recommendations):
    recipe_sim_scores = ingredient_similarities[recipe_index]
    
    # Get the indices of top recommendations
    top_indices = recipe_sim_scores.argsort()[::-1][:num_recommendations]
    top_recommendations = sampled_df.iloc[top_indices][['Name', 'RecipeCategory', 'RecipeIngredientParts']]
    
    return top_recommendations
```

```python
# Assuming you want recommendations for the recipe at index 0 and you want 5 recommendations
recipe_index = 150
num_recommendations = 5

# Get the top recommendations
top_recommendations = text_based_recommendations(recipe_index, num_recommendations)

# Print the top recommendations
print(top_recommendations)
```
